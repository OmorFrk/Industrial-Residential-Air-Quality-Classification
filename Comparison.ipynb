```python
# Comparing Model Accuracies
import matplotlib.pyplot as plt
Logistic_Regression = 97.396
KNN = 98.497
SVM = 97.548
Kernel_SVM = 98.846
Naive_Bayes = 92.349
Decision_Tree = 99.02
Random_Forest = 99.324
models = ['Logistic Regression', 'KNN', 'SVM', 'Kernel SVM', 'Naive Bayes', 'Decision Tree', 'Random Forest']
scores = [Logistic_Regression, KNN, SVM, Kernel_SVM, Naive_Bayes, Decision_Tree, Random_Forest]
blue_shades = ['#c6dbef', '#9ecae1', '#6baed6', '#4292c6', '#2171b5', '#08519c', '#08306b']
plt.figure(figsize=(10, 6))
plt.rcParams['font.family'] = 'Comic Sans MS'
bars = plt.bar(models, scores, color=blue_shades)
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracies')
plt.ylim(90, 100)
plt.xticks(rotation=30)

for bar, score in zip(bars, scores):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.3, f'{score:.2f}%', 
             ha='center', va='bottom', fontsize=10)

plt.show()
```
